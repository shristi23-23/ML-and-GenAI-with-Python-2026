#### Create a function to print first 10 natural numbers

```python
def natural_num(n):
    for i in range(1, n+1):
        print(i, ',' end=" ")

natural_num(10)
```

#### Create a function to calculate sum of first N natural numbers

```python
def sum_natural_num(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum

n = int(input("Enter a number: "))
result = sum_natural_num(n)
print("Sum of first", n, "natural numbers =", result)
```

#### Create a function to reverse a number

```python
def reverse_number(n):
    ans = 0
    while(n > 0):
        digit = n % 10
        ans = (ans * 10) + digit
        n = n//10
    return ans

n = int(input("Enter a number: "))
result = reverse_number(n)
print("Before Reverse: ", n)
print("After Reverse: ", result)
```

#### Create a function to count digits in a number

```python
def count_digits(n):
    count = 0
    while(n > 0):
        n = n//10
        count += 1
    return count

n = int(input("Enter a number: "))
result = count_digits(n)
print("Number: ", n)
print("Number of digits: ", result)
```

#### Create a function to check Palindrome number

```python
def check_palindrome(n):
    original = n
    ans = 0
    while(n > 0):
        digit = n % 10
        ans = (ans * 10) + digit
        n = n//10
    if(ans == original):
        print("Palindrome Number")
    else:
        print("Not a Palindrome Number")

n = int(input("Enter a number: "))
check_palindrome(n)
```

#### Create a function to generate fibonacci series

```python
def fibonacci_series(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci_series(10)
```

#### Calculator using functions that contains the following features :
     -> User selects operation
     -> Program performs calculation
     -> Display Result

```python
def calculator(operation):
    if(operation == "Add"):
            a = int(input("Enter first number:" ))
            b = int(input("Enter secound number: "))
            sum = a + b
            print(a, "+", b, "=", sum)
    elif(operation == "Subtract"):
            a = int(input("Enter first number:" ))
            b = int(input("Enter secound number: "))
            diff = a - b
            print(a, "-", b, "=", diff)
    elif(operation == "Multiply"):
            a = int(input("Enter first number:" ))
            b = int(input("Enter secound number: "))
            prod = a * b
            print(a, "*", b, "=", prod)
    elif(operation == "Divide"):
            a = int(input("Enter first number:" ))
            b = int(input("Enter secound number: "))
            div = a/b
            print(a, "/", b, "=", div)  

print("Choose operation: ")
print ("-> Add")
print("-> Subtract")
print("-> Multiply")
print("-> Divide")

operation = input("Choose operation: ")

print("Operation :-",operation)
print("Result :-", end=" ")
calculator(operation)
```

#### Handle division by zero using exception handling

```python
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))

try:
    div = a/b
    print(div)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed")
```

#### Create a student class with name and marks

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Student Details")
        print("Name: ", self.name)
        print("Marks: ", self.marks)

name = input("Enter student name: ")
marks = input("Enter student marks: ")

s1 = Student(name, marks)

s1.display()
```
